# 1.Rename all the columns by replacing space '', with '_'.

```python

df.rename(columns={'Inspection Result':'Inspection_Result','Inspection Date':'Inspection_Date','Record ID':'Record_ID',
                  'Certificate Number':'Certificate_Number','Building Number':'Building_Number','Business Name':'Business_Name'}, inplace=True)
```

```python
df.rename(columns={'Unit Type':'Unit_Type','Street 2':'Street_2'}, inplace=True)
```

```python
df.columns
```

```python
df['Inspection_Date']=pd.to_datetime(df.Inspection_Date)
df.Inspection_Date.dt.year.head()
```

# 2. Want to know how many total null values are there in  each column

```python
df.isnull().sum()
```

# 3. Show the dataframe containing datas of Inspection Result that was Failed in BROOKLYN City.

```python
df[(df.Inspection_Result=='Fail') & (df.City=='BROOKLYN')].head()
```

# 4. In 2019, on their inspection result, how many people Passed ( in percentage)?

```python
df_2019=df.Inspection_Date.dt.year==2019
df_2019.head()
```

```python
df_2019.value_counts(normalize=True)*100
```

# 5. Which city had the highest number of inspections and how many inspections were conducted in that city?

```python
df.City.value_counts(ascending=False)
```
