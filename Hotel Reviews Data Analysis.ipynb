# 3. Which hotel is least liked by people from the United States of America?

```python
# The best way to approach this problem is to first get a new data set where the nationality is only United States.


df.head(3)
```

```python
#Let's first check how the country name is written in the Reviewer-Nationality column.

df.Reviewer_Nationality.unique()
```

It looks like there are unnecessary spaces in the front and end of the name of the countries.

```python
#Therefore the first thing we need to do is remove the unnecessary extra spaces from the front and the back.

df.Reviewer_Nationality=df['Reviewer_Nationality'].str.strip()
```

```python
df_us=df[df.Reviewer_Nationality=="United States of America"]
df_us.head(4)
```

```python
#Let's get the mean review score for each hotel from the above dataframe

df_us.groupby('Hotel_Name').Reviewer_Score.mean().sort_values(ascending=True)
```

Here, we can see that the average rating by people from USA for "Mercure Paris Bastille Saint Antoine" hotel is the least.
